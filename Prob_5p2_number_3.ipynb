```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dx1dt(x1,x2):
    return -6.2 + 1.1*x1 - 0.8*x2

def dx2dt(x1,x2):
    return -13.8 + 2.1*x1 - 1.2*x2

def f_interaction(s, t):
    "numpy function to compute vector derivative"
    x1 = s[0]
    x2 = s[1]
    return np.array([dx1dt(x1,x2), dx2dt(x1,x2)])

x1domain = np.linspace(0,25,30)
x2domain = np.linspace(0,25,30) # chickens

x1,x2 = np.meshgrid(x1domain, x2domain)

s = np.array([x1,x2])

u,v = f_interaction(s, 0) # get dc/dt and df/dt for all the values of c and f.
uvmag = np.sqrt(u**2+v**2) # magnitude of u,v field for scaling
uvmag = np.where(uvmag<1e-10,1,uvmag) # just avoid dividing by zero

plt.quiver(x1,x2,u/uvmag,v/uvmag,pivot='mid')
plt.grid()
```
