# Slope Fields

The focus of this activity is to learn about *Slope* *Fields*. A slope field is a graphical represenations of a differential equation that hints at the nature of the solutions to that equation. As an example, let's consider the differential equation:

$$ \frac{dy}{dx} = y' = - y/5 $$

Note that this differential equation says the the derivative of y WRT x (i.e., $dy/dx$)is equal to `-y/5`. Look at the code in the following cell that displays the slope field for this equation:

```python
x,y = np.meshgrid(np.linspace(-3,3,20),np.linspace(-3,3,20))
u = 1
v = -y/5
plt.quiver(x,y,u,v,pivot='mid')
plt.grid()
```
