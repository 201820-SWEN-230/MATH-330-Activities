# Slope Fields

The focus of this activity is to learn about *Slope* *Fields*. However, since it's also your first activity, we'll learn some basic facts about differential equations along the way. A slope field is a graphical represenations of a differential equation that hints at the nature of the solutions to that equation. As an example, let's consider the differential equation:

$$ \frac{dy}{dx} = y' = - y/5 $$

Note that this differential equation says the the derivative of y WRT x (i.e., $dy/dx$) is always equal to $-y/5$ independent of $x$. This kind of equation can arise in many different situtations, maybe a mechanical system with fluid flow, or a discharging capacitor, or a chemical reaction relaxing toward equilibrium. $y$ in those cases could be the position of the fluid surface, or the voltage on a capacitor, or the concentration of a reagent. In any case, they all end up with the same differential equation, and they all have the same slope fields! Look at the code in the following cell that displays the slope field for this equation:

```python
xdomain = np.linspace(-3,3,20)
ydomain = np.linspace(-3,3,20)

x,y = np.meshgrid(xdomain, ydomain)

u = 1
v = -y/5

plt.quiver(x,y,u,v,pivot='mid')
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Slope Field for $dy/dx = -y/5$")
```

Let's break this down piece by piece. First look at `np.linspace`:

```python
xdomain = np.linspace(-3,3,20)
xdomain
```

Note that when you have a code cell, the last expression in the code cell is rendered and displayed on the output cell. In this case the last expression is just a variable name: `xdomain` which is the output of `np.linspace` that consists of an array that goes from -3 to 3 with 20 values. Similarly `ydomain` is another array that also goes from -3 to 3 with 20 values:

```python
ydomain = np.linspace(-3,3,20)
ydomain
```

Next, `np.meshgrid` takes these two 20 element 1-D domains and creates *two* 2-D 20x20 arrays with the corresponding $x$ values and $y$ values from the individual domains. This is exactly what's needed for the *Slope* *Field* that shows the slope at every combination of $x$ nd $y$ from the original domains. The output of the `meshgrid` command is a tuple of these two 2-D arrays as you can see below.

```python
x,y = np.meshgrid(xdomain, ydomain)
x
```

```python
y
```

As you can see the variables $x$ and $y$ in this cell are arrays with the values of $x$ and $y$ in every element of the corresponding array.

Next we need to calculate the direction of the derivative at each pair of $x$ and $y$ values. These directions (called $u$ and $v$) are passed into the `quiver` function to display the slope fields. If we make the $x$ component of the direction vector ($u$) equal to 1, then the y component ($v$) should be equal to $dy/dx$, but in this example $dy/dx$ is equal to $-y/5$ so you can see that we're just setting $v=-y/5$.

Finally, we should find the actual analytical solution to the differential equation and plot that on the slope field so we can see how it compares.

Now, because this equation is pretty simple it's easy to see that:

$$dy/dx = -y/5 $$

$$\int dy/y = -\int dx/5$$

so:

$$ \ln(y) = -\frac{x}{5} + C $$

or equivalently:

$$ y(x) = A e^{-x/5} $$

Where $A$ is just a constant that happens to be equal to $e^C$. $A$ is a parameter of the general solution. It could have any value, and no matter what value it has, the relationships $y(x)$ is still a solution to the original differential equation. Usually there will be some "initial" value for $y$ at some particular value of $x$ that constrains $A$ to have a particular value. For example, if we knew that $y(0)=2$ then $A$ would have to have a value of 2 so that $2e^{0/5}$ would be 2.

```python
xdomain = np.linspace(-3,3,20)
ydomain = np.linspace(-3,3,20)

x,y = np.meshgrid(xdomain, ydomain)

u = 1
v = -y/5

plt.quiver(x,y,u,v,pivot='mid')

y_analytical = 2*np.exp(-xdomain/5)
plt.plot(xdomain, y_analytical, label="$y(0)=2$")

plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Slope Field plus solution")
plt.legend()
```

Actually, you can see that $y(0)=A$ so we can generate any solution that starts at $y(0)$ by using:

$$y(x) = y(0) e^{-y/5}$$

Let's graph those on top of the slope field! Note that since the solutions extend above $y=\pm 3$ we need to adjust the domain limits and the figure size to keep the aspect ratio of the figure the same as the domain. If you don't do this the arrows will not point in the correct direction!

```python
xdomain = np.linspace(-3,3,20)
ydomain = np.linspace(-6,6,40)

x,y = np.meshgrid(xdomain, ydomain)

u = 1
v = -y/5

plt.figure(figsize=(5,10))
plt.quiver(x,y,u,v,pivot='mid')
plt.plot(xdomain, 3*np.exp(-xdomain/5),label="$y(0)=3$")
plt.plot(xdomain, 2*np.exp(-xdomain/5),label="$y(0)=2$")
plt.plot(xdomain, 1*np.exp(-xdomain/5),label="$y(0)=1$")
plt.plot(xdomain, -1*np.exp(-xdomain/5),label="$y(0)=-1$")
plt.plot(xdomain, -2*np.exp(-xdomain/5),label="$y(0)=-2$")
plt.plot(xdomain, -3*np.exp(-xdomain/5),label="$y(0)=-3$")
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Slope Field plus solution")
plt.legend()
```

Notice how the slope field guides the eye to suggest the behavior of the solutions. Note also that python, being a general purpose programming language, also has things like loops! This means we don't have to manually plot each line as a separate instruction, but we can write a loop to do that:

```python
xdomain = np.linspace(-3,3,20)
ydomain = np.linspace(-6,6,40)

x,y = np.meshgrid(xdomain, ydomain)

u = 1
v = -y/5

plt.figure(figsize=(5,10))
plt.quiver(x,y,u,v,pivot='mid')

for y0 in [-3,-2,-1,1,2,3]: # note this loop is over the initial conditions
    plt.plot(xdomain, y0*np.exp(-xdomain/5),label=f"$y(0)={y0}$")

plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Slope Field plus solution")
plt.legend()
```

Let's model the amount of medicine currently in the body as a variable called $y$. Through research and clinical trials you know that the human body metabolizes (eliminates) the medication at a rate that's proportaional to the amount of drug currently in the body. Let's call the rate of metabolization $m$ (units of "per hour"). We need to translate this real world scenario into a differential equation that we can use to make predictions about how much drug we expect to see in the body at an time after the patch is applied. We know the drug is introduced into the body at a steady rate by the patch, but that it's taken out at a rate proportional to the current value of $y$ with the proportionality constant being $m {\rm \ hr^{-1}}$. This gives us a model of:

$$ \frac{dy}{dt} = r - m y $$

Where $r$ is the constant rate the drug is introduced, and $m$ is the proportionality constant for the rate at which the drug is metabolized.
