# Slope Fields

The focus of this activity is to learn about *Slope* *Fields*. A slope field is a graphical represenations of a differential equation that hints at the nature of the solutions to that equation. As an example, let's consider the differential equation:

$$ \frac{dy}{dx} = y' = - y/5 $$

Note that this differential equation says the the derivative of y WRT x (i.e., $dy/dx$) is always equal to $-y/5$ independent of $x$. Look at the code in the following cell that displays the slope field for this equation:

```python
xdomain = np.linspace(-3,3,20)
ydomain = np.linspace(-3,3,20)

x,y = np.meshgrid(xdomain, ydomain)

u = 1
v = -y/5

plt.quiver(x,y,u,v,pivot='mid')
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Slope Field for $dy/dx = -y/5$")
```

Let's break this down piece by piece. First look at `np.linspace`:

```python
xdomain = np.linspace(-3,3,20)
xdomain
```

Note that when you have a code cell, the last expression in the code cell is rendered and displayed on the output cell. In this case the last expression is just a variable name: `xdomain` which is the output of `np.linspace` that consists of an array that goes from -3 to 3 with 20 values. Similarly `ydomain` is another array that also goes from -3 to 3 with 20 values:

```python
ydomain = np.linspace(-3,3,20)
ydomain
```

Next, `np.meshgrid` takes these two 20 element 1-D domains and creates *two* 2-D 20x20 arrays with the corresponding $x$ values and $y$ values from the individual domains. This is exactly what's needed for the *Slope* *Field* that shows the slope at every combination of $x$ nd $y$ from the original domains.

```python
x,y = np.meshgrid(xdomain, ydomain)
x
```
